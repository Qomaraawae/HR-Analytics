```python
# ========================================
# SEL 1: KONFIGURASI PATH & IMPORT LIBRARY
# ========================================

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')
import joblib

# ========================================
# KONFIGURASI PATH
# ========================================

BASE_PATH = '/Users/macbookpro/Documents/UAD/MATKUL/INTELEJEN BISNIS/HR_Analytics/'

# Buat semua folder jika belum ada
folders = ['Data', 'Models', 'Images']
for folder in folders:
    os.makedirs(os.path.join(BASE_PATH, folder), exist_ok=True)
    print(f"✅ Folder {folder} siap")

# Path lengkap
DATA_SOURCE = os.path.join(BASE_PATH, 'HR_Analytics.xlsx')
DATA_FOLDER = os.path.join(BASE_PATH, 'Data')
MODELS_FOLDER = os.path.join(BASE_PATH, 'Models')
IMAGES_FOLDER = os.path.join(BASE_PATH, 'Images')

print(f"\n📁 Base Path: {BASE_PATH}")
print(f"📁 Data Folder: {DATA_FOLDER}")
print(f"📁 Models Folder: {MODELS_FOLDER}")
print(f"📁 Images Folder: {IMAGES_FOLDER}")

# ========================================
# SETTING VISUALISASI
# ========================================

try:
    plt.style.use('seaborn-v0_8-whitegrid')
except:
    try:
        plt.style.use('seaborn-whitegrid')
    except:
        plt.style.use('default')

sns.set_palette("Set2")
print("\n✅ Semua library dan konfigurasi berhasil!")
```

```python
# ========================================
# SEL 2: LOAD DATA
# ========================================

df = pd.read_excel(DATA_SOURCE, sheet_name='Sheet1')

print("="*60)
print("INFORMASI DATA")
print("="*60)
print(f"Jumlah data: {df.shape[0]} baris, {df.shape[1]} kolom")
print("\n5 Data Pertama:")
display(df.head())

print("\nTipe Data per Kolom:")
print(df.dtypes)

print("\nStatistik Deskriptif:")
display(df.describe())

print("\nCek Missing Values:")
print(df.isnull().sum())

print("="*60)
print("✅ Data berhasil diload!")
```

```python
# ========================================
# SEL 3: PREPROCESSING DATA
# ========================================

# 1. Hapus kolom yang tidak diperlukan
kolom_dihapus = ['EmployeeCount', 'Over18', 'StandardHours']
df_clean = df.drop(columns=kolom_dihapus)
print("✅ Kolom yang dihapus:", kolom_dihapus)

# 2. Perbaiki nama kolom (BOM)
df_clean.columns = df_clean.columns.str.replace('ï»¿', '')
df_clean.columns = df_clean.columns.str.replace('¬ª', '')
df_clean.columns = df_clean.columns.str.replace('¬ø', '')
print("✅ Nama kolom telah dibersihkan")

# 3. Encode kolom kategorikal
le = LabelEncoder()
kolom_kategorikal = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 
                     'Gender', 'JobRole', 'MaritalStatus', 'OverTime']

for col in kolom_kategorikal:
    df_clean[col + '_Encoded'] = le.fit_transform(df_clean[col])

print("✅ Kolom kategorikal berhasil di-encode")
print(f"   - {', '.join(kolom_kategorikal)}")

# 4. Tampilkan hasil
print("\nData setelah preprocessing:")
print(f"Jumlah data: {df_clean.shape[0]} baris, {df_clean.shape[1]} kolom")
display(df_clean.head())

# 5. Simpan ke folder Data
df_clean.to_csv(os.path.join(DATA_FOLDER, 'HR_Analytics_Clean.csv'), index=False)
print(f"\n✅ Data bersih disimpan di: {DATA_FOLDER}/HR_Analytics_Clean.csv")
```

```python
# ========================================
# SEL 4: EDA - EKSPLORASI VISUALISASI
# ========================================

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('HUMAN RESOURCES ANALYTICS - EDA VISUALIZATION', fontsize=16, fontweight='bold')

# 1. Distribusi Attrition
df_clean['Attrition'].value_counts().plot(kind='bar', ax=axes[0,0], color=['green','red'])
axes[0,0].set_title('1. Distribusi Attrition')
axes[0,0].set_xlabel('Attrition (0=Tidak, 1=Ya)')
axes[0,0].set_ylabel('Jumlah Karyawan')

# 2. Attrition per Department
pd.crosstab(df_clean['Department'], df_clean['Attrition']).plot(kind='bar', ax=axes[0,1])
axes[0,1].set_title('2. Attrition per Department')
axes[0,1].set_xlabel('Department')
axes[0,1].set_ylabel('Jumlah Karyawan')
axes[0,1].legend(['Tidak', 'Ya'])

# 3. Distribusi Gaji
df_clean['MonthlyIncome'].hist(bins=30, ax=axes[0,2], color='skyblue', edgecolor='black')
axes[0,2].set_title('3. Distribusi Monthly Income')
axes[0,2].set_xlabel('Monthly Income (USD)')
axes[0,2].set_ylabel('Frekuensi')

# 4. Attrition berdasarkan Overtime
pd.crosstab(df_clean['OverTime'], df_clean['Attrition']).plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title('4. Attrition vs OverTime')
axes[1,0].set_xlabel('OverTime (No/Yes)')
axes[1,0].set_ylabel('Jumlah Karyawan')
axes[1,0].legend(['Tidak', 'Ya'])

# 5. Job Satisfaction vs Attrition
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df_clean, ax=axes[1,1])
axes[1,1].set_title('5. Job Satisfaction vs Attrition')
axes[1,1].set_xlabel('Attrition')
axes[1,1].set_ylabel('Job Satisfaction (1-4)')

# 6. YearsAtCompany vs Attrition
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df_clean, ax=axes[1,2])
axes[1,2].set_title('6. Years At Company vs Attrition')
axes[1,2].set_xlabel('Attrition')
axes[1,2].set_ylabel('Years At Company')

plt.tight_layout()
plt.savefig(os.path.join(IMAGES_FOLDER, 'EDA_Visualization.png'), dpi=300, bbox_inches='tight')
plt.show()

print(f"✅ Visualisasi EDA disimpan di: {IMAGES_FOLDER}/EDA_Visualization.png")
```

```python
# ========================================
# SEL 5: PERSIAPAN DATA UNTUK DATA MINING
# ========================================

# Pilih fitur numerik
fitur_numerik = ['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction',
                 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome',
                 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating',
                 'RelationshipSatisfaction', 'StockOptionLevel', 'TotalWorkingYears',
                 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany',
                 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']

# Fitur encoded
fitur_encoded = ['Attrition_Encoded', 'BusinessTravel_Encoded', 'Department_Encoded', 
                 'EducationField_Encoded', 'Gender_Encoded', 'JobRole_Encoded', 
                 'MaritalStatus_Encoded', 'OverTime_Encoded']

# Gabungkan semua fitur
X_full = df_clean[fitur_numerik + fitur_encoded]

# Pisahkan untuk klasifikasi (Attrition)
y_attrition = df_clean['Attrition_Encoded']
X_class = X_full.drop(columns=['MonthlyIncome'])

# Pisahkan untuk regresi (MonthlyIncome)
y_income = df_clean['MonthlyIncome']
X_reg = X_full.drop(columns=['MonthlyIncome'])

# Standarisasi untuk Clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_full)

print("="*60)
print("DATA SIAP UNTUK MODELING")
print("="*60)
print(f"✅ Total fitur: {X_full.shape[1]}")
print(f"✅ Fitur Klasifikasi: {X_class.shape[1]} fitur")
print(f"✅ Fitur Regresi: {X_reg.shape[1]} fitur")
print(f"✅ Fitur Clustering: {X_scaled.shape[1]} fitur")
```

```python
# ========================================
# SEL 6: KLASIFIKASI - PREDIKSI ATTRITION
# ========================================

print("="*60)
print("MODEL KLASIFIKASI: PREDIKSI ATTRITION")
print("="*60)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_class, y_attrition, test_size=0.2, random_state=42, stratify=y_attrition
)

# Model Random Forest
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X_train, y_train)

# Prediksi
y_pred = rf_clf.predict(X_test)

# Evaluasi
print("\n📊 HASIL EVALUASI:")
print(f"Akurasi: {accuracy_score(y_test, y_pred):.3f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Tidak Attrition', 'Attrition']))

# Feature Importance
importance = pd.DataFrame({
    'Feature': X_class.columns,
    'Importance': rf_clf.feature_importances_
}).sort_values('Importance', ascending=False)

print("\n🏆 TOP 10 FITUR PALING BERPENGARUH:")
print(importance.head(10).to_string(index=False))

# Visualisasi Feature Importance
plt.figure(figsize=(10, 6))
plt.barh(importance.head(10)['Feature'], importance.head(10)['Importance'], color='coral')
plt.xlabel('Nilai Importance')
plt.title('Top 10 Fitur - Prediksi Attrition')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig(os.path.join(IMAGES_FOLDER, 'Feature_Importance_Attrition.png'), dpi=300)
plt.show()

# Simpan model ke folder Models
joblib.dump(rf_clf, os.path.join(MODELS_FOLDER, 'attrition_model.pkl'))
print(f"\n✅ Model klasifikasi disimpan di: {MODELS_FOLDER}/attrition_model.pkl")
```

```python
# ========================================
# SEL 7: REGRESI - PREDIKSI GAJI
# ========================================

print("="*60)
print("MODEL REGRESI: PREDIKSI MONTHLY INCOME")
print("="*60)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_reg, y_income, test_size=0.2, random_state=42
)

# Model Random Forest Regressor
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(X_train, y_train)

# Prediksi
y_pred = rf_reg.predict(X_test)

# Evaluasi
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\n📊 HASIL EVALUASI:")
print(f"R-squared (R²): {r2:.3f}")
print(f"RMSE: ${rmse:,.2f}")
print(f"MSE: ${mse:,.2f}")

# Feature Importance untuk Regresi
importance_reg = pd.DataFrame({
    'Feature': X_reg.columns,
    'Importance': rf_reg.feature_importances_
}).sort_values('Importance', ascending=False)

print("\n🏆 TOP 10 FITUR PALING BERPENGARUH UNTUK GAJI:")
print(importance_reg.head(10).to_string(index=False))

# Visualisasi
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Income')
plt.ylabel('Predicted Income')
plt.title(f'Actual vs Predicted Income (R² = {r2:.3f})')
plt.tight_layout()
plt.savefig(os.path.join(IMAGES_FOLDER, 'Regression_Results.png'), dpi=300)
plt.show()

# Simpan model ke folder Models
joblib.dump(rf_reg, os.path.join(MODELS_FOLDER, 'income_model.pkl'))
print(f"\n✅ Model regresi disimpan di: {MODELS_FOLDER}/income_model.pkl")
```

```python
# ========================================
# SEL 8: CLUSTERING - SEGMENTASI KARYAWAN
# ========================================

print("="*60)
print("CLUSTERING: SEGMENTASI KARYAWAN")
print("="*60)

# Pilih fitur untuk clustering
fitur_cluster = ['MonthlyIncome', 'YearsAtCompany', 'TotalWorkingYears', 
                 'Age', 'JobSatisfaction']

# Standarisasi data
scaler_cluster = StandardScaler()
X_cluster_scaled = scaler_cluster.fit_transform(df_clean[fitur_cluster])

# Metode Elbow untuk menentukan K optimal
inertias = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_cluster_scaled)
    inertias.append(kmeans.inertia_)

# Plot Elbow
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.axvline(x=4, color='r', linestyle='--', label='K=4 (Optimal)')
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(IMAGES_FOLDER, 'Elbow_Method.png'), dpi=300)
plt.show()

print("✅ Berdasarkan Elbow Method, K=4 dipilih sebagai optimal")

# Latih K-Means dengan K=4
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df_clean['Cluster'] = kmeans.fit_predict(X_cluster_scaled)

# Analisis setiap cluster
print("\n📊 ANALISIS KARAKTERISTIK SETIAP CLUSTER:")
cluster_summary = df_clean.groupby('Cluster')[fitur_cluster].mean()
display(cluster_summary)

# Visualisasi Cluster
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: MonthlyIncome vs YearsAtCompany
scatter1 = axes[0].scatter(df_clean['YearsAtCompany'], df_clean['MonthlyIncome'], 
                          c=df_clean['Cluster'], cmap='viridis', alpha=0.6)
axes[0].set_xlabel('Years At Company')
axes[0].set_ylabel('Monthly Income')
axes[0].set_title('Cluster: Years vs Income')
plt.colorbar(scatter1, ax=axes[0])

# Plot 2: Age vs JobSatisfaction
scatter2 = axes[1].scatter(df_clean['Age'], df_clean['JobSatisfaction'], 
                          c=df_clean['Cluster'], cmap='viridis', alpha=0.6)
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Job Satisfaction')
axes[1].set_title('Cluster: Age vs Job Satisfaction')
plt.colorbar(scatter2, ax=axes[1])

plt.tight_layout()
plt.savefig(os.path.join(IMAGES_FOLDER, 'Clustering_Results.png'), dpi=300)
plt.show()

# Interpretasi Cluster
print("\n📋 INTERPRETASI CLUSTER:")
for i in range(4):
    cluster_data = df_clean[df_clean['Cluster'] == i]
    print(f"\n🔹 Cluster {i} ({len(cluster_data)} karyawan):")
    print(f"   - Rata-rata Gaji: ${cluster_data['MonthlyIncome'].mean():,.0f}")
    print(f"   - Rata-rata Lama Kerja: {cluster_data['YearsAtCompany'].mean():.1f} tahun")
    print(f"   - Rata-rata Usia: {cluster_data['Age'].mean():.1f} tahun")
    print(f"   - Job Satisfaction: {cluster_data['JobSatisfaction'].mean():.2f}")
    attrition_rate = cluster_data['Attrition'].value_counts(normalize=True).get('Yes', 0)*100
    print(f"   - Attrition Rate: {attrition_rate:.1f}%")
```

```python
# ========================================
# SEL 9: IDENTIFIKASI KARYAWAN BERISIKO & EKSPOR DATA
# ========================================

print("="*60)
print("IDENTIFIKASI KARYAWAN BERISIKO TINGGI")
print("="*60)

# Prediksi probabilitas attrition untuk semua karyawan
df_clean['Attrition_Probability'] = rf_clf.predict_proba(X_class)[:, 1]

# Ambil 10 karyawan dengan risiko tertinggi
top_risk = df_clean.nlargest(10, 'Attrition_Probability')

# Kolom yang ditampilkan
kolom_tampilan = ['EmpID', 'Age', 'Department', 'JobRole', 'MonthlyIncome', 
                  'YearsAtCompany', 'JobSatisfaction', 'OverTime', 'Attrition_Probability']

# Filter kolom yang tersedia
kolom_tersedia = [col for col in kolom_tampilan if col in top_risk.columns]
top_risk_display = top_risk[kolom_tersedia]

print("\n🔴 10 KARYAWAN DENGAN RISIKO ATTRITION TERTINGGI:")
display(top_risk_display)

# Simpan ke folder Data
top_risk_display.to_csv(os.path.join(DATA_FOLDER, 'Top_Risk_Employees.csv'), index=False)
print(f"\n✅ Daftar karyawan berisiko disimpan di: {DATA_FOLDER}/Top_Risk_Employees.csv")

# Simpan data untuk dashboard
df_clean.to_csv(os.path.join(DATA_FOLDER, 'hr_data_for_dashboard.csv'), index=False)
print(f"✅ Data untuk dashboard disimpan di: {DATA_FOLDER}/hr_data_for_dashboard.csv")

# Rekomendasi
print("\n💡 REKOMENDASI:")
print("1. Karyawan dengan probabilitas > 70% harus segera diintervensi")
print("2. Fokus pada faktor-faktor utama: OverTime, JobSatisfaction, dan YearsAtCompany")
print("3. Pertimbangkan program retensi khusus untuk karyawan berisiko tinggi")
```

```python
# ========================================
# SEL 12: RINGKASAN & REKOMENDASI BISNIS (VERSI FINAL - TANPA ERROR)
# ========================================

print("="*60)
print("RINGKASAN ANALISIS & REKOMENDASI BISNIS")
print("="*60)

print("\n📌 RINGKASAN DATA MINING:")

# ========================================
# 1. KLASIFIKASI - Gunakan variabel yang sudah disimpan
# ========================================

print("\n1️⃣ KLASIFIKASI (Prediksi Attrition)")

# Ambil metrik dari model yang sudah dilatih
try:
    # Gunakan akurasi dari hasil training di Sel 6
    # Kita simpan akurasi di variabel terpisah
    from sklearn.metrics import accuracy_score
    
    # Cek apakah variabel y_test_class dan y_pred_class ada
    if 'y_test' in dir() and 'y_pred' in dir():
        # Tapi ini bisa jadi dari regresi, jadi kita hitung ulang
        # Gunakan data training yang sudah ada
        X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
            X_class, y_attrition, test_size=0.2, random_state=42, stratify=y_attrition
        )
        y_pred_clf = rf_clf.predict(X_test_clf)
        akurasi = accuracy_score(y_test_clf, y_pred_clf)
        print(f"   - Akurasi Model: {akurasi:.1%}")
    else:
        print("   - ⚠️ Metrik tidak tersedia")
except Exception as e:
    print(f"   - ⚠️ Error: {str(e)[:50]}...")

# Feature Importance
try:
    if 'importance' in dir():
        print(f"   - Top 3 Fitur Penyebab Attrition:")
        for i in range(min(3, len(importance))):
            print(f"     • {importance.iloc[i]['Feature']}: {importance.iloc[i]['Importance']:.3f}")
    else:
        print("   - ⚠️ Feature importance tidak tersedia")
except:
    print("   - ⚠️ Feature importance tidak tersedia")

# ========================================
# 2. REGRESI
# ========================================

print("\n2️⃣ REGRESI (Prediksi Gaji)")

try:
    # Gunakan data regresi
    X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
        X_reg, y_income, test_size=0.2, random_state=42
    )
    y_pred_reg = rf_reg.predict(X_test_reg)
    
    from sklearn.metrics import r2_score, mean_squared_error
    r2 = r2_score(y_test_reg, y_pred_reg)
    rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
    
    print(f"   - R-squared (R²): {r2:.3f}")
    print(f"   - RMSE: ${rmse:,.2f}")
except Exception as e:
    print(f"   - ⚠️ Error: {str(e)[:50]}...")

try:
    if 'importance_reg' in dir():
        print(f"   - Top 3 Fitur Penentu Gaji:")
        for i in range(min(3, len(importance_reg))):
            print(f"     • {importance_reg.iloc[i]['Feature']}: {importance_reg.iloc[i]['Importance']:.3f}")
    else:
        print("   - ⚠️ Feature importance tidak tersedia")
except:
    print("   - ⚠️ Feature importance tidak tersedia")

# ========================================
# 3. CLUSTERING
# ========================================

print("\n3️⃣ CLUSTERING (Segmentasi Karyawan)")

try:
    if 'Cluster' in df_clean.columns:
        for i in range(4):
            cluster_data = df_clean[df_clean['Cluster'] == i]
            if len(cluster_data) > 0:
                print(f"   - Cluster {i}: {len(cluster_data)} karyawan")
                print(f"     • Rata-rata Gaji: ${cluster_data['MonthlyIncome'].mean():,.0f}")
                attrition_rate = cluster_data['Attrition'].value_counts(normalize=True).get('Yes', 0)*100
                print(f"     • Attrition Rate: {attrition_rate:.1f}%")
    else:
        print("   - ⚠️ Cluster belum dibuat")
except:
    print("   - ⚠️ Data cluster tidak tersedia")

# ========================================
# REKOMENDASI
# ========================================

print("\n" + "="*60)
print("💡 REKOMENDASI STRATEGIS:")
print("="*60)

print("\n🔴 PRIORITAS TINGGI:")
print("1. Intervensi terhadap karyawan dengan probabilitas attrition > 70%")
print("2. Kurangi jam kerja lembur (OverTime) untuk mengurangi attrition")
print("3. Tingkatkan Job Satisfaction melalui program employee engagement")

print("\n🟡 PRIORITAS SEDANG:")
print("4. Perhatikan karyawan di Cluster 3 (underpaid, masa kerja panjang)")
print("5. Program mentoring untuk karyawan baru (YearsAtCompany < 2)")
print("6. Evaluasi kebijakan kompensasi untuk karyawan berkinerja tinggi")

print("\n🟢 PRIORITAS RENDAH:")
print("7. Monitoring rutin untuk karyawan dengan risiko moderate")
print("8. Pengembangan karir untuk karyawan di Cluster 2")
print("9. Program retensi untuk karyawan senior (Cluster 1)")

print("\n" + "="*60)
print("✅ ANALISIS SELESAI!")
print("="*60)
```
